```python
# Load packages
import numpy as np

from toygrad.nn.datasets.load import load_MNIST
from toygrad.nn.models import FCNetwork
from toygrad.nn.optim import ADAM
```

### Load data

```python
# Load MNIST data
data = load_MNIST()
train, validation, test = data

# Plot some examples?
```

### Initialize Network

```python
# Initialize Network
np.random.seed(0)
layers_size = [28*28, 30, 10]
net = FCNetwork(layers_size)
```

### Training

```python
# Training parameters
epoch_tol = 8 # Max number of epochs we tolerate without improving val_accuracy
batch_size = 20
# SGD
#init_lr, momentum, weight_decay = 1e-2, 0.9, 1e-3
#optimizer = SGD(net.parameters, init_lr, momentum, weight_decay)
# ADAM
init_lr = 1e-2
optimizer = ADAM(net.parameters, init_lr)
save_model = 'MNIST_model'

# Train model
out = net.train_model(data, batch_size, optimizer, epoch_tol, save_model)  
```
