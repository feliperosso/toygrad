```python
# Load packages
import numpy as np

from toygrad.nn.datasets.load import load_MNIST
from toygrad.nn.models import FCNetwork
from toygrad.nn.optim import ADAM
import matplotlib.pyplot as plt
```

### Load data

```python
# Load MNIST data
data = load_MNIST()
train, validation, test = data

# Plot an example
rnd = 43
image = train[0][rnd].reshape(28, 28)
label = train[1][rnd]
plt.imshow(image, cmap='gray')
plt.title(f'Label: {label}')
plt.show()
```

### Initialize Network

```python
# Initialize Network
np.random.seed(0)
layers_size = [28*28, 30, 10]
dropout_rate = 0.1
net = FCNetwork(layers_size, dropout_rate)
```

### Training

```python
# Training parameters
epoch_tol = 4 # Max number of epochs we tolerate without improving val_accuracy
batch_size = 20
# ADAM optimizer
init_lr = 1e-2
optimizer = ADAM(net.parameters, init_lr)
save_model = 'MNIST_model'

# Train model
out = net.train_model(data, batch_size, optimizer, epoch_tol, save_model)  
```
