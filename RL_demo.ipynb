```python
import numpy as np
import gymnasium as gym
from toygrad.nn.models import VanillaRL
from toygrad.nn.optim import ADAM
```

Cart Pole

```python
np.random.seed(0)
# Initialize policy and value networks
env = gym.make("CartPole-v1")
vanilla_rl = VanillaRL(env)

# Policy training parameters
batch_size = 20
policy_epochs = 100
policy_lr = 5e-3 
lmbda = 1.0
policy_optimizer = ADAM(vanilla_rl.parameters, policy_lr, maximize=True)

# Value estimator training parameters
value_epochs = 10
value_net_lr = 1e-2
value_optimizer = ADAM(vanilla_rl.value_net.parameters, value_net_lr)

# Train
out = vanilla_rl.train_policy(batch_size, policy_epochs, value_epochs, policy_optimizer, value_optimizer, lmbda)
epoch_rewards, smooth_reward = out
```
